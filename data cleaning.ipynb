```python
from enum import Enum

import numpy as np
import pandas as pd

River = Enum(
    "River",
    [
        "YAMUNA",
        "SATLUJ",
        "BEAS",
        "GANGA",
        "MAHANADI",
        "BRAHMAPUTRA",
        "GODAVARI",
        "KRISHNA",
        "CAUVERY",
    ],
)


cols = [
    "CODE",
    "LOCATION",
    "STATE",
    "T_MIN",
    "T_MAX",
    "T_AVG",
    "DO_MIN",
    "DO_MAX",
    "DO_AVG",
    "PH_MIN",
    "PH_MAX",
    "PH_AVG",
    "COND_MIN",
    "COND_MAX",
    "COND_AVG",
    "BOD_MIN",
    "BOD_MAX",
    "BOD_AVG",
    "N_MIN",
    "N_MAX",
    "N_AVG",
    "FC_MIN",
    "FC_MAX",
    "FC_AVG",
    "TC_MIN",
    "TC_MAX",
    "TC_AVG",
]
```

```python
def process(file_name, *args):
    table = pd.concat(args, axis=0)
    table.columns = cols
    table.reset_index(inplace=True, drop=True)
    table.to_pickle(f"./{file_name}.pkl")
    return table


def preprocess(x, i, j, row_drop, col_drop):
    x = (
        (
            x.drop(axis=0, index=range(i))
            .replace("BDL", 0)
            .replace("_", np.nan)
            .replace("-", np.nan)
        )
        .iloc[:, :j]
        .drop(axis=0, index=row_drop)
        .drop(axis=1, columns=col_drop)
    )
    x.columns = range(19)
    return x


def table(
    excel_name, sheet_name, skip_header, skip_last_cols, row_drop=[], col_drop=[]
):
    return list(
        map(
            lambda i, x: preprocess(
                x,
                skip_header[i],
                skip_last_cols[i],
                row_drop[i] if len(row_drop) > 0 else [],
                col_drop[i] if len(col_drop) > 0 else [],
            ),
            range(len(sheet_name)),
            list(
                pd.read_excel(excel_name, sheet_name=sheet_name, header=None).values()
            ),
        )
    )
```

```python
import os
from pathlib import Path

import numpy as np

data_path = Path("data/pkl/")
files = list(
    map(lambda x: x.name.replace(".pkl", "").split("_"), data_path.glob("*.pkl"))
)
river2year = {}
year2river = {}
for file in files:
    if file[0] in river2year:
        river2year[file[0]].append(int(file[1]))
    else:
        river2year[file[0]] = [int(file[1])]
    if int(file[1]) in year2river:
        year2river[int(file[1])].append(file[0])
    else:
        year2river[int(file[1])] = [file[0]]

df = pd.DataFrame(columns=river2year.keys(), index=range(2012, 2022))
for river in river2year.keys():
    df[river] = list(map(lambda x: river2year[river].count(x), list(df[river].index)))
```

```python
for y in range(2013, 2014):
    paths = {k: Path(f"data/pkl/{k}_{y}.pkl") for k in year2river[y]}
    df_y = pd.DataFrame()
    for river, path in paths.items():
        df = pd.read_pickle(path)
        df["RIVER"] = river
        df = df[["RIVER", *df.columns[:-1]]]
        df_y = pd.concat([df_y, df], axis=0)
    df_y.to_csv(f"data/csv/{y}.csv", index=False)
```
